```python
def fig1(energies,diff_array,std_array,MAES):
    import numpy as np
    import matplotlib.pyplot as plt
    import seaborn as sns
    import matplotlib.ticker as mticker
    from matplotlib.transforms import offset_copy
    
    #sns.set_style("whitegrid")
    plt.rc('text',color='k')
    plt.rcParams['font.family'] = 'sans-serif'
    
    
    
    fig,axes = plt.subplots(4,3,figsize=(15/1.5,2*7.5/1.5),dpi=200,sharex=False)
    fig.tight_layout(pad=3)
    rows = ['A', 'B', 'C', 'D']
    xloc = [0.,0.,0,0]
    count=0
    for ax, row in zip(axes[:,0], rows):
        ax.annotate(row, xy=(xloc[count], 0.5), xytext=(-ax.yaxis.labelpad - 5, 0),
                    xycoords=ax.yaxis.label, textcoords='offset points',
                    size=15, ha='right', va='center',weight='bold')
        count += 1
    
    
    axes = axes.flatten('F')
    
    ##########Distribution#############
    import matplotlib.pylab as pl
    color = pl.cm.viridis(np.linspace(0,1,5))
    #print(color.shape)
    palette=list(plt.cm.viridis(np.linspace(0, 1, 5)))
    #color=np.asarray(['tab:blue','tab:red','purple','darkgreen','orange'])
    axes[0].set_title('Benzene',fontsize=25)#,fontweight="bold")
    axes[4].set_title('Naphthalene',fontsize=25)#,fontweight="bold")
    axes[8].set_title('Anthracene',fontsize=25)#,fontweight="bold")
    
    count=0
    
    for i in [0,4,8]:
        #axes[i].sharey(axes[0])
        axes[i].set_xlabel(r'$E^f-\bar{E}^f$ [eV]',fontsize=15)
        #axes[i].grid(visible=None)
        for j in range(4,-1,-1):
            sns.kdeplot(ax=axes[i],data=energies[count][j],
                        common_norm=True, common_grid=True,
                        bw_method='scott',levels=10, thresh=0.05, 
                        gridsize=200, cut=3, #palette=palette[4-j],
                        color=color[4-j], 
                        fill=False,bw_adjust=1,
                        cumulative=False)#
        count += 1
    axes[8].legend(['TZVP','SVP','6-31G','3-21G','STO-3G'],frameon=True,framealpha=0.0,fontsize=7,ncol=1,bbox_to_anchor=(1.042,1))
    
    axes[0].set_ylabel('Distribution',fontsize=15)
    axes[4].set_ylabel('',fontsize=15)
    axes[8].set_ylabel('',fontsize=15)
    
    
    ###########Scatter#############
    import matplotlib.pylab as pl
    colors = pl.cm.viridis(np.linspace(0,1,5))
    #color=np.asarray(['tab:red','purple','darkgreen','orange'])
    markers = ['^','*','P','d']
    stride = [16,8,4,2]
    count = 0
    axes[1].set_ylabel(r'$E^f-\bar{E}^f$ [eV]',fontsize=15)
    for k in [1,5,9]:
        
        axes[k].set_xlabel(r'E(TZVP)-$\bar{E}$(TZVP) [eV]',fontsize=15)
        #axes[k].grid(visible=None)
        for j in range(0,4):
            axes[k].scatter(energies[count][-1],energies[count][3-j][0::stride[3-j]],s=1,color=colors[j+1],marker=markers[j])
        count += 1
    axes[9].legend(['SVP','6-31G','3-21G','STO-3G'],frameon=True,framealpha=1.0,fontsize=7,markerscale=5)
    
    ###########Hierarchy####################
    count = 0
    xlabels = ['STO-3G','3-21G','6-31G','SVP']
    axes[2].set_ylabel('$\Delta E_f^{\mathrm{TZVP}}$ [eV]',fontsize=15)
    for i in [2,6,10]:
        #axes[i].grid(visible=None)    
        axes[i].errorbar(xlabels,diff_array[count],std_array[count],color='red',ecolor='k',capsize=3,elinewidth=1,barsabove=True)
        count += 1
        #axes[i].set_yscale('log')
        axes[i].sharey(axes[6])
    
    ######SingleFidelity-LC#################
    n = 2**np.arange(1,10)
    count = 0
    formatter = mticker.ScalarFormatter()
    formatter.set_scientific(False)
    axes[3].set_ylabel('MAE [meV]',fontsize=15)
    
    y_minor = mticker.LogLocator(base = 0.1, subs = np.arange(1.0, 8.0,2), numticks = 10)
    
    
    for i in [3,7,11]:    
        axes[i].sharey(axes[3])
        axes[i].sharex(axes[3])
        
        axes[i].loglog(n,1e3*MAES[count],marker='o',color=color[0],linestyle='--')
        
        axes[i].set_xlabel('$N_{\mathrm{train}}^{\mathrm{TZVP}}$',fontsize=15)
        
        axes[i].grid('minor',axis='y',ls='--',color='lightgray')
        axes[i].grid('both','major',ls='-',color='gray')
        #axes[i].yaxis.set_minor_formatter(formatter)
        axes[i].yaxis.set_major_formatter(formatter)
        axes[i].yaxis.set_major_locator(y_minor)
        count += 1
    axes[3].xaxis.set_major_formatter(mticker.ScalarFormatter())
    axes[3].set_xticks([2,8,32,128,512])
    #################################
    for ax in axes:
        ax.tick_params(axis='x', labelsize=12)
        ax.tick_params(axis='y', labelsize=12)
```

```python
def MD_energy_diffs(molecule):
    strides = [16,8,4,2]
    energies = np.load(f'Data/energyarrs/{molecule}_MD_energy_array.npy',allow_pickle=True)
    
    en_diffs = []
    diff_stds = []
    for i in range(4):
        en_diffs.append(np.mean(np.abs(energies[i][0::strides[i],1]-energies[-1][:,1])))
        diff_stds.append(np.std(np.abs(energies[i][0::strides[i],1]-energies[-1][:,1])))
    
    for i in range(5):
        energies[i] = energies[i][:,1]
    
    return energies,en_diffs,diff_stds
    
```

```python
def DFTB_energy_diffs(molecule):
    strides = [16,8,4,2]
    energies = np.load(f'Data/energyarrs/{molecule}_DFTB_energy_array.npy',allow_pickle=True)
    
    en_diffs = []
    diff_stds = []
    for i in range(4):
        en_diffs.append(np.mean(np.abs(energies[i][0::strides[i],1]-energies[-1][:,1])))
        diff_stds.append(np.std(np.abs(energies[i][0::strides[i],1]-energies[-1][:,1])))
    
    for i in range(5):
        energies[i] = energies[i][:,1]
    
    return energies,en_diffs,diff_stds
    
```

```python
molecules = np.asarray(['benzene','naphthalene','anthracene'])
cmd_ens = np.zeros((3),dtype=object)
cmd_diff = np.zeros((3),dtype=object)
cmd_diffstd = np.zeros((3),dtype=object)

dftb_ens = np.zeros((3),dtype=object)
dftb_diff = np.zeros((3),dtype=object)
dftb_diffstd = np.zeros((3),dtype=object)

for i in range(0,3):
    cmd_ens[i],cmd_diff[i],cmd_diffstd[i] = MD_energy_diffs(molecules[i])
    dftb_ens[i],dftb_diff[i],dftb_diffstd[i] = DFTB_energy_diffs(molecules[i])
```

```python
def combined_scatter(outputs,true_energies):
    import numpy as np
    import matplotlib.pyplot as plt
    plt.style.use('default')
    #plt.rc('text', usetex=False)
    predicted = outputs['Predictions'] #shape[5,2]

    fig, ax = plt.subplots(outputs['n'].shape[0], 2, figsize = (8,16), sharex=True)
    fig.tight_layout(pad=2.5)
    KRRtitles = ['$P^{TZVP}_{KRR}$','$P^{SVP}_{KRR}$',
                 '$P^{6-31G}_{KRR}$','$P^{3-21G}_{KRR}$','$P^{STO-3G}_{KRR}$']
    MFMLtitles = ['$P^{TZVP}_{KRR}$','$P^{(TZVP;SVP)}_{MFML}$',
                 '$P^{(TZVP;6-31G)}_{MFML}$','$P^{(TZVP;3-21G)}_{MFML}$','$P^{(TZVP;STO-3G)}_{MFML}$']
    all_titles = np.vstack((KRRtitles,MFMLtitles))
    color=color = pyl.cm.viridis(np.linspace(0,1,5))
    markers = ['o','^','*','P','d']
    
    m = ['Baseline','MFML']
    aa = ['TZVP','SVP','6-31G','3-21G','STO-3G']

    for i in range(0,outputs['n'].shape[0]):
        for j in range(0,2):
            ax[i][j].plot([np.min(true_energies)-np.std(true_energies),np.max(true_energies)+np.std(true_energies)],
                          [np.min(true_energies)-np.std(true_energies),np.max(true_energies)+np.std(true_energies)],
                          '--',color='k', linewidth=0.75)
            ax[i][j].scatter(true_energies,predicted[outputs['n'].shape[0]-1-i][j], 
                             s=1, color=color[i],marker=markers[i])
            
            ax[i][j].set_title(all_titles[j][i],fontsize=15,color='k')
            ax[i][j].text(x=np.mean(true_energies)-2.5*np.std(true_energies),y=np.max(true_energies),
                          s='MAE: '+str(np.round(np.mean(np.abs(predicted[outputs['n'].shape[0]-1-i][j]-true_energies)),4))+
                          '(eV)',fontsize=15,color='darkslategrey')

            if j==0:
                ax[i][j].set_ylabel('KRR Prediction (eV)', fontsize=15,color='k')
            else:
                ax[i][j].set_ylabel('MFML Prediction (eV)', fontsize=15,color='k')

    ax[i][j-1].set_xlabel('Reference Energy (eV)', fontsize=15, color='k')
    ax[i][j].set_xlabel('Reference Energy (eV)', fontsize=15,color='k')
    
    #return fig
```

```python
def level_errors(outputs,true_vals,x_low=None,x_up=None,ax=None):
    import numpy as np
    import matplotlib.pyplot as plt
    import matplotlib.ticker as mticker
    plt.style.use('default')
    
    base_pred = np.copy(outputs['Predictions'][:,0])
    MFML_pred = np.copy(outputs['Predictions'][:,1])

    delta_preds= []
    
    for i in range(3,-1,-1):
        temp = MFML_pred[i]-base_pred[i] - MFML_pred[i+1]+base_pred[i+1]
        delta_preds.append(temp)
    
    delta_preds = np.asarray(delta_preds)
    delta_preds = np.copy(delta_preds[::-1])
    
    ax = ax or plt.gca()
    ax.plot(1e3*np.abs(true_vals-base_pred[1]),label='$P_{KRR}^{(3-21G)}$',color='navy',linewidth=1.5)
    
    p_temp1 = np.copy(base_pred[0])+np.copy(delta_preds[0])
    tt = np.abs(true_vals-p_temp1)
    ax.plot(1e3*tt,label='$P_{KRR}^{(STO-3G)}+P_{KRR}^{(STO-3G,3-21G)}$',color='salmon',linestyle='-',linewidth=1.5)
    
    ax.grid('both','both',color='gray', linestyle='--', linewidth=0.5)
    ax.set_xlim(x_low,x_up)
    #ax.set_yscale('log')
    return tt
```
